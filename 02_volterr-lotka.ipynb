## Model Volterra - Lotki

```python

%matplotlib inline
import matplotlib.pyplot as plt

import numpy
from scipy.integrate import odeint

def rhs(y, t):
    
    a = 0.6 # prawdopodobieństwo narodzenia królika
    c = 0.0051 # prawdopodobieństwo zabicia królika przez lisa
    b = 1.7 # prawdopodobieństwo śmierci lisa
    p1 = y[0]
    p2 = y[1] 
    dp1dt = a * p1 - c * p1 * p2 # króliki
    dp2dt = c * p1 * p2 - b * p2 # lisy
    return numpy.array([ dp1dt , dp2dt ])

p0 = numpy.array([70, 50]) # liczebność początkowa króliki,lisy
ts = numpy.arange(0, 30, 0.1) # wirualny okres czasu: 0 - 30 (co 0.1)
res = odeint( rhs, p0, ts )#compute solution
p1 = res[:, 0] #extract p1 (króliki)
p2 = res[:, 1] #p2 (lisy)

import matplotlib.pyplot as plt
plt.plot(ts, p1,'-b', label='kroliki')
plt.plot(ts, p2,'-g',label='lisy')
plt.legend()
plt.xlabel('t')
plt.show()

plt.plot(p1, p2,'r-')
plt.xlabel('kroliki')
plt.ylabel('lisy')
plt.show()
```
