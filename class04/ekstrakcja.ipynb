### Przykład - jednostopniowa ekstrakcja równowagowa

#### Dane:
K = x4/x2 = 4 - współczynnik podziału

F1 = 60 kg/h

F3 = 40 kg/h

x1s = 0.01 - stężenie ekstahowanego surowca w wodzie

x3r = 0 - do ekstrakcji uzywany jest czysty rozpuszczalnik organiczny

s = surowiec, r = rozpuszczanik, w = woda
#### Założenia: 

1. W aparacie ustala się równowaga

2. Rozpuszczalnik nie miesza się z wodą, nie zanieczyszcza rafinatu

3. Woda nie zanieczyszcza ekstraktu

#### Obliczyć x2, F2, x4, F4.

```python
# spróbować samodzielnie stworzyć model, a następnie obliczyć x2,F2, x4,F4
#
# w razie problemów, poniżej jest rozwiązanie ...



































```

```python
from scipy.optimize import fsolve

# parametry    
K = 2.0
F1 = 60 #kg/h
F3 = 40 #kg/h
x1s = 0.01
x3r = 1.0 # czysty rozpuszczalnik
    
# model
def ekstrakcja(X):
   

    
    # niewiadome
    x2s, F2, x4s, F4 = X
    
    # --- UKŁAD RÓWNAŃ
    
    # bilans masowy
    eq1 = F1 + F3 - F4 - F2
    
    # z założenia o ustaleniu się stanu równowagi podczas ekstrakcji
    eq2 = x4s - K*x2s
    
    # bilans składnika ekstrahowanego
    eq3 = F1*x1s - F4*x4s -F2*x2s
    
    # bilans rozpuszczalnika (patrz założenia)
    eq4 = F3*x3r - F4*(1-x4s)
    
    return [eq1, eq2, eq3, eq4]


# poeksperymentować z guess ...
guess = [0, 0, 0, 0]

x2, F2, x4, F4 = fsolve(ekstrakcja, guess)

# raport = wyniki
print("Wynik: x2={0:.4f} F2={1:.2f} x4={2:.4f} F4={3:.2f}".format(x2, F2, x4, F4))

# sprawdzenie obliczeń ...

print(F1+F3)
print(F2+F4)

```
