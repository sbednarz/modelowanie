# Model Lotki-Volterry

Model matematyczny:

$$\frac{dx}{dt}=ax-bxy$$
$$\frac{dy}{dt}=cxy-dy$$

a,b,c,d - stałe (por. kod poniżej).

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/9/91/Wykresruchomy.gif/220px-Wykresruchomy.gif">

```python
# Tutaj zaczyna się
# skrypt w Pythonie = model matematyczny + wizualizacja symulacji
#

%matplotlib inline
import matplotlib.pyplot as plt

import numpy
from scipy.integrate import odeint

def model(y, t):
    
    # Parametry modelu => zmieniamy/eksperymentujemy/obserwujemy zachowanie modelu
    a = 0.1   # prawdopodobieństwo narodzenia królika
    b = 0.02  # prawdopodobieństwo zabicia królika przez lisa
    c = 0.01  # prawdopodobieństwo narodzin lisa 
    d = 0.2   # prawdopodobieństwo śmierci lisa
    
    p1 = y[0]
    p2 = y[1] 
    
    # układ równań różnuczkowych
    dp1dt = a * p1 - b * p1 * p2 # króliki
    dp2dt = c * p1 * p2 - d * p2 # lisy
    return numpy.array([ dp1dt , dp2dt ])


p0 = numpy.array([50, 10]) # liczebność początkowa królików (50), lisy (10)
ts = numpy.arange(0, 200, 0.1) # wirualny okres czasu: 0 - 200 (krok 0.1)

res = odeint( model, p0, ts ) # symulacja, całkowanie modelu po czasie ts, dla warunków początkowych p0

p1 = res[:, 0] #extract p1 (króliki)
p2 = res[:, 1] #p2 (lisy)


import matplotlib.pyplot as plt
plt.plot(ts, p1,'-b', label='kroliki')
plt.plot(ts, p2,'-g',label='lisy')
plt.legend()
plt.xlabel('t')
plt.show()

plt.plot(p1, p2,'r-')
plt.xlabel('kroliki')
plt.ylabel('lisy')
plt.show()
```
