```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit


# dane pomiarowe:
xdata = [0,   1,   2, 5, 10, 22]
ydata = [0, 2.1, 3.2, 13,21, 42]


# model matematyczny (czasem typu czarnej skrzynki)

def model(x, a, b): # równanie do dopasowania + parametry
    return a * x + b


p, pcov = curve_fit(model , xdata, ydata)

a, b = p

# dane
plt.plot(xdata, ydata, 'bo', label='dane')

# model
x = np.linspace(0, 25, 50)
plt.plot(x, model(x, a, b), '-', label='fit')

# umieszczenie na wykresie wartości estymowanych parametró
plt.title('(a,b)=%s' % p)
plt.legend()
plt.xlabel('x')
plt.ylabel('y')

print a,b


```

### Przykład 2. Dopasować do danych funkcję $y = ax + \frac{b}{x} + c$

```python
xdata = [100,120,200,250]
ydata = [0.197506482, 0.184596634, 0.254637152, 0.298805538]
```

```python
# model matematyczny (czasem typu czarnej skrzynki)

def model(x, a, b, c): # równanie do dopasowania + parametry
    return a * x + b/x + c


p, pcov = curve_fit(model , xdata, ydata)

a, b, c = p

# dane
plt.plot(xdata, ydata, 'bo')

# model
x = np.linspace(100, 250, 50)
plt.plot(x, model(x, a, b, c), '-')

# umieszczenie na wykresie wartości estymowanych parametró
plt.title('(a,b)=%s' % p)
plt.xlabel('x')
plt.ylabel('y')

print a,b,c
```
